```python
# Splitting the image into 3 parts

top = img[:160,:,1] * 2
mid = img[160:320,:,2] * 2
botm = img[320:,:,0] * 0.5

final_img = np.zeros((480,640))

final_img[:160,:] = top * 255
final_img[160:320,:] = mid * 255
final_img[320:,:] = botm * 255

plt.imshow(final_img,cmap='gray')
```

```python
cv2.imwrite('decoded_watson_img.png', final_img)
```
